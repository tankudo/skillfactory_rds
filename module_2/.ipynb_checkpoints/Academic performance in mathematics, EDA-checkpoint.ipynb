**Описание датасета** (переменные, которые содержит датасет):
1. <span style="background-color:#a0ffa7; ">school</span> — аббревиатура школы, в которой учится ученик
2. <span style="background-color:#a0ffa7; ">sex</span> — пол ученика ('F' - женский, 'M' - мужской)
3. <span style="background-color:#a0ffa7; ">age</span> — возраст ученика (от 15 до 22)
4. <span style="background-color:#a0ffa7; ">address</span> — тип адреса ученика ('U' - городской, 'R' - за городом)
5. <span style="background-color:#a0ffa7; ">famsize</span> — размер семьи('LE3' <= 3, 'GT3' >3)
6. <span style="background-color:#a0ffa7; ">Pstatus</span> — статус совместного жилья родителей ('T' - живут вместе 'A' - раздельно)
7. <span style="background-color:#a0ffa7; ">Medu</span> — образование матери (0 - нет, 1 - 4 класса, 2 - 5-9 классы, 3 - среднее специальное или 11 классов, 4 - высшее)
8. <span style="background-color:#a0ffa7; ">Fedu</span> — образование отца (0 - нет, 1 - 4 класса, 2 - 5-9 классы, 3 - среднее специальное или 11 классов, 4 - высшее)
9. <span style="background-color:#a0ffa7; ">Mjob</span> — работа матери ('teacher' - учитель, 'health' - сфера здравоохранения, 'services' - гос служба, 'at_home' - не работает, 'other' - другое)
10. <span style="background-color:#a0ffa7; ">Fjob</span> — работа отца ('teacher' - учитель, 'health' - сфера здравоохранения, 'services' - гос служба, 'at_home' - не работает, 'other' - другое)
11. <span style="background-color:#a0ffa7; ">reason</span> — причина выбора школы ('home' - близость к дому, 'reputation' - репутация школы, 'course' - образовательная программа, 'other' - другое)
12. <span style="background-color:#a0ffa7; ">guardian</span> — опекун ('mother' - мать, 'father' - отец, 'other' - другое)
13. <span style="background-color:#a0ffa7; ">traveltime</span> — время в пути до школы (1 - <15 мин., 2 - 15-30 мин., 3 - 30-60 мин., 4 - >60 мин.)
14. <span style="background-color:#a0ffa7; ">studytime</span> — время на учёбу помимо школы в неделю (1 - <2 часов, 2 - 2-5 часов, 3 - 5-10 часов, 4 - >10 часов)
15. <span style="background-color:#a0ffa7; ">failures</span> — количество внеучебных неудач (n, если 1<=n<3, иначе 4)
16. <span style="background-color:#a0ffa7; ">schoolsup</span> — дополнительная образовательная поддержка (yes или no)
17. <span style="background-color:#a0ffa7; ">famsup</span> — семейная образовательная поддержка (yes или no)
18. <span style="background-color:#a0ffa7; ">paid</span> — дополнительные платные занятия по математике (yes или no)
19. <span style="background-color:#a0ffa7; ">activities</span> — дополнительные внеучебные занятия (yes или no)
20. <span style="background-color:#a0ffa7; ">nursery</span> — посещал детский сад (yes или no)
21. <span style="background-color:#a0ffa7; ">higher</span> — хочет получить высшее образование (yes или no)
22. <span style="background-color:#a0ffa7; ">internet</span> — наличие интернета дома (yes или no)
23. <span style="background-color:#a0ffa7; ">romantic</span> — в романтических отношениях (yes или no)
24. <span style="background-color:#a0ffa7; ">famrel</span> — семейные отношения (от 1 - очень плохо до 5 - очень хорошо)
25. <span style="background-color:#a0ffa7; ">freetime</span> — свободное время после школы (от 1 - очень мало до 5 - очень мого)
26. <span style="background-color:#a0ffa7; ">goout</span> — проведение времени с друзьями (от 1 - очень мало до 5 - очень много)
27. <span style="background-color:#a0ffa7; ">health</span> — текущее состояние здоровья (от 1 - очень плохо до 5 - очень хорошо)
27. <span style="background-color:#a0ffa7; ">absences</span> — количество пропущенных занятий
29. <span style="background-color:#a0ffa7; ">score</span> — баллы по госэкзамену по математике

```python
stud.loc[(stud.traveltime == 1) & (stud.reason == 'None'), 'reason'] = "home"
pd.DataFrame(stud.reason.value_counts())
```

Выброс — значение -1. Скорее всего оно равно 1 (или у кого-то хорошее чувство юмора, или случайно попал знак минус). Заменим его. В том числе и отсутствующие значения модой, так как отношения у подавляющего большинства на четверку.

```python
stud.famrel.fillna(stud.famrel.mode()[0], inplace=True)
stud.loc[(stud['famrel'] == -1), 'famrel'] = 1
stud.famrel.hist(color='green', bins=9)
print(
    f'Количество пропущенных значений в столбце famrel = {stud.famrel.isnull().sum()}'
)
plt.title('farmel')
```

absences — количество пропущенных занятий

```python
print(
    f'Количество пропущенных значений в столбце absences = {stud.absences.isnull().sum()}'
)
stud.absences.hist(color='green', bins=70)
stud.absences.describe()
plt.title('absences')
```

Есть несколько выбросов. Воспользуемся формулой интеркватрального расстояния.

```python
print(
    f'Количество пропущенных значений в столбце absences = {stud.absences.isnull().sum()}'
)
median = stud.absences.median()
IQR = stud.absences.quantile(0.75) - stud.absences.quantile(0.25)
perc25 = stud.absences.quantile(0.25)
perc75 = stud.absences.quantile(0.75)
print(
    '25-й перцентиль: {},'.format(perc25),
    '75-й перцентиль: {},'.format(perc75), "IQR: {}, ".format(IQR),
    "Границы выбросов: [{f}, {l}].".format(f=perc25 - 1.5 * IQR,
                                           l=perc75 + 1.5 * IQR))
stud.absences.loc[stud.absences.between(perc25 - 1.5 * IQR, perc75 +
                                        1.5 * IQR)].hist(bins=20,
                                                         range=(0, 20),
                                                         label='IQR',
                                                         color='green')
plt.title('absences')
```

Этот способ позволил отобрать значения пропусков экстремально большие. Отфильтруем эти данные.

```python
stud = stud.loc[stud.absences.between(perc25 - 1.5 * IQR, perc75 + 1.5 * IQR)]
```

<span style="background-color:green; color:white; font-family: 'Bebas Neue'; font-size: 3em">score</span>

score — баллы по госэкзамену по математике

```python
print(
    f'Количество пропущенных значений в столбце absences = {stud.score.isnull().sum()}'
)
stud.score.hist(color='green', bins=70)
stud.score.describe()
plt.title('score')
```

Это наша целевая переменная. Распределение нормальное, если не считать учеников, получивших 0 баллов.
Пропусков всего 6, заменим их на среднее значение

```python
stud.score = stud.score.fillna(stud.score.mean())
print('Количество пропущенных значений', stud.score.isnull().sum())
```

Нулевая оценка, по моему мнению, получена в результате нарушения, которое может быть не связана с успеваемостью. Например, не пришел на экзамен. Потому. Создадим еще один столбец 'no_score' в котором "1" присвоим нулевому баллу на экзамене, и "0", если оценка отличается от нулевой. После этого нулевые оценки заменим на среднее значение.

```python
stud['no_score'] = stud.score.apply(lambda x: 1 if x == 0 else 0)
stud.no_score.value_counts()
```

Теперь заменим нулевые оценки на средние значения.

```python
stud.score = stud.score.apply(lambda x: stud.score.mean() if x == 0 else x)
stud.score.hist(color='green', bins=60)
plt.title('score')
```

```python
def get_boxplot(column):
    fig, ax = plt.subplots(figsize=(14, 4))
    sns.boxplot(x=column,
                y='score',
                data=stud.loc[stud.loc[:, column].isin(
                    stud.loc[:, column].value_counts().index[:10])],
                ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot for ' + column)
    plt.show()
```

```python
for col in ['school', 'sex', 'address', 'famsize', 'Pstatus',
       'Mjob', 'Fjob', 'reason', 'guardian', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
       'higher', 'internet', 'romantic', 'famrel', 'goout']:
    get_boxplot(col)
```

```python
def get_stat_dif(column):
    cols = stud.loc[:, column].value_counts().index[:]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        ttest = ttest_ind(
            stud.loc[stud.loc[:, column] == comb[0], 'score'].dropna(),
            stud.loc[stud.loc[:, column] == comb[1], 'score'].dropna()).pvalue
        #         print(f"для столбца {column} ttest:= {ttest}")
        #         print(f"                     comb:= {combinations_all}")
        if ttest <= 0.05 / len(combinations_all):  # Учли поправку Бонферони
            print('!!!Найдены статистически значимые различия для колонки',
                  column)
            break
```

```python
def get_stat_dif(column):
    cols = stud.loc[:, column].value_counts().index[:]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        ttest = ttest_ind(
            stud.loc[stud.loc[:, column] == comb[0], 'no_score'].dropna(),
            stud.loc[stud.loc[:, column] == comb[1],
                     'no_score'].dropna()).pvalue
        #         print(f"для столбца {column} ttest:= {ttest}")
        #         print(f"                     comb:= {combinations_all}")
        if ttest <= 0.05 / len(combinations_all):  # Учли поправку Бонферони
            print('!!!Найдены статистически значимые различия для колонки',
                  column)
            break
```

- В столбце Medu поменяли на значение верхнего соседа.

Гипотезы:

- положительная корреляция параметра studytime и score может говорить о том, что чем больше занимается ученик, тем лучше учится.
- отрицательная корреляция параметра failures и score может говорить о том, что чем больше неудач по другим предметам тем ниже score. И положительная корреляция в случае не получения оценки. Больше неудач, отсутствие попыток слать экзамены. Самая значимая величина на оценку в таблице.
- отрицательная корреляция параметра goout и score может говорить о том, что чем больше ученик проводит времени с друзьями тем ниже score
- положительная корреляция по параметру Medu говорит о том, что чем выше лучше образование матери тем выше score
