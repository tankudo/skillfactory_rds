```python
print(
    f'     absences = {stud.absences.isnull().sum()}'
)
median = stud.absences.median()
IQR = stud.absences.quantile(0.75) - stud.absences.quantile(0.25)
perc25 = stud.absences.quantile(0.25)
perc75 = stud.absences.quantile(0.75)
print(
    '25- : {},'.format(perc25),
    '75- : {},'.format(perc75), "IQR: {}, ".format(IQR),
    " : [{f}, {l}].".format(f=perc25 - 1.5 * IQR,
                                           l=perc75 + 1.5 * IQR))
stud.absences.loc[stud.absences.between(perc25 - 1.5 * IQR, perc75 +
                                        1.5 * IQR)].hist(bins=20,
                                                         range=(0, 20),
                                                         label='IQR',
                                                         color='green')
plt.legend()
```

```python
def get_boxplot(column):
    fig, ax = plt.subplots(figsize = (14, 4))
    sns.boxplot(x=column, y='score', 
                data=stud.loc[stud.loc[:, column].isin(stud.loc[:, column])],
               ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot for ' + column)
    plt.show()
```

```python
for col in stud:
    get_boxplot(col)
```
